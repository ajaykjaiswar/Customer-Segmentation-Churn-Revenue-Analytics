# Data Analysis
## Customer Segmentation

```python
# reference date to find recency as data is old
referencedate = transactions["Transaction_Date"].max() + pd.Timedelta(days=1)
```

```python
referencedate
```

```python
df = pd.read_sql("""Select CustomerID, current_date(),
                    max(Transaction_Date) as Recent from transactions
                    group by CustomerID
                    order by Recent desc""",con=engine)
```

```python
rfm=transactions.groupby("CustomerID").agg({"Tenure_Months":"max",
                                            "Transaction_Date": lambda x : (referencedate - x.max()).days,
                                           "Transaction_ID": "count",
                                           "total_amt": "sum"})
```

```python
rfm.columns
```

```python
rfm.columns=["Tenure_months","Recency","Frequency","Monetary"]
```

```python
rfm.describe()
```

```python
# Quantile-Based Scoring

rfm["R_score"]=pd.qcut(rfm["Recency"],5,labels=[5,4,3,2,1])
rfm["F_score"]=pd.qcut(rfm["Frequency"].rank(method="first"),5,labels=[1,2,3,4,5])  
rfm["M_score"]=pd.qcut(rfm["Monetary"],5,labels=[1,2,3,4,5])
```

```python
rfm.describe()
```

```python
def rfm_segment(row):
    if row["R_score"] >= 4 and row["F_score"] >= 4 and row["M_score"] >=4:
        return "High Value"
    elif row["R_score"] >= 3 and row["F_score"] >= 3:
        return "Loyal"
    elif row["R_score"] >= 4 and row["F_score"] <= 2:
        return "Potential"
    elif row["R_score"] <= 2 and row["F_score"] >= 3:
        return "At Risk"
    else:
        return "Lost"
 
```

```python
rfm["Segment"]=rfm.apply(rfm_segment,axis=1)
```

```python
rfm
```

```python
rfm.reset_index(inplace=True)
rfm.columns
```

```python
rfm.to_sql("rfm_analysis",con=engine,if_exists="replace",index=False)
```

## Churn Rate by Segment

```python
def churn(row):
    if row["Recency"]>90:
        return 1
    else:
        return 0
```

```python
rfm["Churned"]=rfm.apply(churn,axis=1)
rfm
```

```python
segment_analysis = rfm.groupby("Segment").agg(Customers=("Segment", "count"),
                            Avg_Frequency=("Frequency", "mean"),
                            Total_Revenue=("Monetary", "sum"),
                            Churned_Customers=("Churned", "sum"))
segment_analysis
```

```python
segment_analysis["Customer_Share_%"] =(100*segment_analysis["Customers"]/segment_analysis["Customers"].sum()).round(2)
segment_analysis["Churn_Rate_%"] = (100*segment_analysis["Churned_Customers"]/segment_analysis["Customers"]).round(2)
segment_analysis["Churn_Share_%"]=(100*segment_analysis["Churned_Customers"]/segment_analysis["Churned_Customers"].sum()).round(2)
segment_analysis["Revenue_Share_%"] =(100*segment_analysis["Total_Revenue"]/segment_analysis["Total_Revenue"].sum()).round(2)
```

```python
segment_analysis
```

```python
# Reorder Rows

segment_order = ["High Value", "Loyal", "Potential", "At Risk", "Lost"]
segment_analysis = segment_analysis.reindex(segment_order)

# Reorder Columns

column_order = ["Customers",
                "Customer_Share_%",
                "Total_Revenue",
                "Revenue_Share_%",
                "Avg_Frequency",
                "Churned_Customers",
                "Churn_Rate_%",
                "Churn_Share_%"]
segment_analysis = segment_analysis[column_order]
```

```python
segment_analysis
```

```python
plot_df = segment_analysis.loc[segment_order].reset_index()
sns.set_style("white")

fig, ax1 = plt.subplots(figsize=(10, 6))

# Primary Y-axis â†’ Churn Share
sns.barplot(
    data=plot_df,
    x="Segment",
    y="Churn_Share_%",
    ax=ax1,
    color="tab:red",
    alpha=0.6,
    label="Churn Share (%)"    
)

ax1.set_ylabel("Churn Share (%)")
ax1.set_xlabel("Customer Segment")


# Secondary Y-axis â†’ Revenue Share
ax2 = ax1.twinx()

sns.lineplot(
    data=plot_df,
    x="Segment",
    y="Revenue_Share_%",
    ax=ax2,
    color="tab:blue",
    marker="o",
    label="Revenue Share (%)",
    legend=False
)

ax2.set_ylabel("Revenue Share (%)")

h1, l1 = ax1.get_legend_handles_labels()
h2, l2 = ax2.get_legend_handles_labels()

ax1.legend(
    h1 + h2,
    l1 + l2,
    loc="upper center",
    frameon=False
)

plt.title("Churn vs Revenue Contribution by Customer Segment")
plt.tight_layout()
plt.show()
```

## Hypothesis Testing: Churn differences across segments

**Research Question:**  
Is there a significant difference in Churn Rate across different segments?

### Hypotheses

- **Hâ‚€ (Null Hypothesis):**  
  Churn rate is the same across customer segments.

- **Hâ‚ (Alternative Hypothesis):**  
  Churn rate differs across customer segments.


```python
contingency = pd.crosstab(
    rfm["Segment"],
    rfm["Churned"])

chi2, p_value, dof, expected = chi2_contingency(contingency)

p_value
```

```python
if p_value < 0.05:
    print("Reject Hâ‚€: Churn is significantly different across segments")
else:
    print("Fail to Reject Hâ‚€: Churn is same across segments")
```

## Hypothesis Testing: Revenue differences across segments

**Research Question:**  
Is there a significant difference in revenue between churned and retained customers?

### Hypotheses

- **Hâ‚€ (Null Hypothesis):**  
  Mean revenue of churned = mean revenue of retained.

- **Hâ‚ (Alternative Hypothesis):**  
  Mean revenue of churned â‰  mean revenue of retained.

```python
rfm.groupby("Churned")["Monetary"].describe()
```

- Mean >> Median, therefore its a right skewed data 

```python
from scipy.stats import ttest_ind

churned = rfm.loc[rfm["Churned"] == 1, "Monetary"]
retained = rfm.loc[rfm["Churned"] == 0, "Monetary"]

t_stat, p_ttest = ttest_ind(churned, retained, equal_var=False)
p_ttest
```

```python
if p_ttest < 0.05:
    print("Reject Hâ‚€: Mean revenue of churned â‰  mean revenue of retained")
else:
    print("Fail to Reject Hâ‚€: Mean revenue of churned = mean revenue of retained")
```
